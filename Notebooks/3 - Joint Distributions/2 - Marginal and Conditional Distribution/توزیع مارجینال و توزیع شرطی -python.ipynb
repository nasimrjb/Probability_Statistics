        فرض کنید این جدول را به شما داده‌اند و می‌خواهید توزیع نمره دانشجویان را بررسی کنید.
        برای پیدا کردن توزیع نمره دانشجویان، باید احتمال رخ دادن نمره‌ها را بین دانشجویان بررسی کنید. مثلا برای کسب نمره ۲۰ تا۳۹ باید احتمال اینکه دانشجویی این بازه نمره را کسب کند حساب کنید. برای این اتفاق، مهم نیست که دانشجو چند ساعت مطالعه کرده است و فقط نمره او مهم است. پس با جمع کردن، می‌توانیم اثر میزان مطالعه را حذف کنیم.

        توزیع حاشیه‌ای از یک توزیع توام، توزیع هر کدام از متغیر‌های تصادفی توزیع توام را به صورت جداگانه نشان می‌دهد.
        </br>

    در واقع در این توزیع، به شرط دانستن اطلاعاتی راجع به یکی از متغیر‌های تصادفی، می‌خواهیم توزیع دیگری یا توزیع دیگران را بررسی کنیم.

    توزیع حاشیه‌ای(مارجینال)
</font>
		<p></p>
		<hr>
        فرض کنید توزیع توام دو متغیر تصادفی 
        $x$ 
        و
        $y$ 
         به ما داده شده است. 
         گاهی نیاز داریم توزیع مستقل دو متغیر تصادفی را هم از توزیع توام به دست آوریم. 
         جواب این سوال ما در توزیع حاشیه‌ای نهفته است.
         برای حساب کردن توزیع حاشیه‌ای یکی از این متغیر‌ها کافی است به نوعی تاثیر آنرا بر روی تابع توزیع توام حذف کنیم.
         برای این هدف، باید از انتگرال برای توزیع‌های پیوسته و سیگما برای توزیع‌های گسسته روی تمام حالات متغیری که قرار است حذف شود استفاده کنیم. مثلا در توزیع گسسته:
         $$P_Y(y) = \Sigma_{x:p(x,y) > 0} p(x,y) $$
         در توزیع پیوسته، سیگما با انتگرال جایگزین می‌شود: 
         $$f_Y(y) = \int_{-\infty}^{+\infty} f(x,y) dx$$
     <br></br>
      ستون آخر جدول مثال بالا یا سطر آخر آن همین عملیات جمع کردن روی تمام حالت‌های فضا را انجام می‌دهد.
      <br>
      </br>
      حال فرض کنید می‌خواهیم توزیع تجمعی حاشیه‌ای یک متغیر تصادفی را از روی توزیع تجمعی توام بدون استفاده از تابع چگالی آنها بدست آوریم. در این صورت داریم:
$$F_X(x) = F_{X,Y}(x,\inf) = lim_{y\rightarrow \inf} F_{XY}(x,y)$$
         </font>

		<hr>&ensp;&ensp;&ensp;&ensp;
        <font color="red">مثال:</font>
        در این مثال قرار است به سراغ توزیع چند جمله‌ای برویم. فرض کنید سه عدد
        $n\in \mathbb{Z} , p,q \in (0,1)$
        داریم که
        $p+q \leq 1$.
        فرض کنید توزیع توام دو متغیر تصادفی 
        $X$ و 
        $Y$
        به شکل زیر است:
        $$P(X=i,Y=j) = \begin{cases} \frac{n!}{i!j!(n-i-j)!}p^i q^j(1-p-q)^{n-i-j} & i\geq 0, j \geq 0, i+j\leq n\\
        0 & o.w.
        \end{cases}$$
        که 
        $i,j \in \mathbb{N}$
        . حال می‌خواهیم توزیع 
        $X$ 
        را حساب کنیم. داریم:
        $$P(X=i) = \begin{cases} \frac{n!}{i!(n-i)!} p^i \Sigma_{j=0}^{n-i} \frac{(n-i)!}{j!(n-i-j)!} q^j (1-p-q)^{n-i-j} & i=0,1,\cdots,n \\
        0 & i<0 \text{ or } i>n
        \end{cases}$$
        به عنوان تمرین، عبارت داخل سیگما را ساده کنید. در نهایت به عبارت زیر خواهید رسید:
        $$P(X=i) = \begin{cases}{n\choose i} p^i (1-p)^{n-i} & i= 0,1,\cdots n \\
        0 & i>0 \text{ or } i<n\end{cases}$$
        مشاهده می‌کنید که
        به همان توزیع دو جمله‌ای رسیده‌ایم!
        لطفا در این نتیجه کمی تامل کنید!
     </font>
</div>

<div id="sec_conditional_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توزیع شرطی
            (حالت گسسته)
        </font>
		<p></p>
		<hr>
 با مفهوم شرطی کردن پیشامد‌ها قبلا آشنا شدیم. 
        تعریف توزیع‌های شرطی مشابه است. برای توزیع‌های گسسته، تابع احتمال شرطی 
        $X$
        به شرط
        $Y=y$
        برای همه مقادیر 
        $y$
        که 
        $p_Y(y)>0$ 
        است، به صورت زیر تعریف می‌شود:
        $$P_{X|Y}(x|y) = P(X=x|Y=y) = \frac{P(X=x , Y = y)}{P(Y=y)}$$
        $$ = \frac{P(x,y)}{P_Y(y)}$$
        در واقع خود را به فضایی که 
        $Y$
        مقدار خاص
        $y$ 
        را اتخاذ کرده‌است، محدود می‌کنیم.
        </font>
</div>

<br />
<div id="sec_conditional_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
       </font>
		<p></p>
		<hr>&ensp;&ensp;&ensp;&ensp;
        <font color="red">مثال:</font>
فرض کنید 
        $n$
        بار سکه پرتاب کرده‌ایم. 
        فرض کنید 
        $k$
        بار شیر آمده است. می‌خواهیم ثابت کنیم تمام ترتیب‌های ممکن شیر و خط که ممکن است رخ دهد، هم احتمال هستند.
        <br></br>
        حل: برای این کار می‌توان یک رشته تصادفی مانند رشته 
        $o= (h,t,h,t,h,t,..)$
        را در نظر گرفت. 
        حال در نظر بگیرید احتمال اینکه این رشته اتفاق افتاده باشد در حالیکه 
        $k$
        بار شیر آمده است:
        (احتمال شیر آمدن را 
        $p$
        در نظر بگیرید)
        $$ P(o|X=k) = \frac{P(o,X=k)}{P(X=k)} = \frac{P(o)}{P(X=k)}$$
        $$=\frac{p^k (1-p)^{n-k}}{{n \choose k}p^k(1-p)^{n-k}} = \frac{1}{{n\choose k}}$$
        این یکی از 
        ${n\choose k}$
        حالت ممکن بود که احتمال آن با هر حالت دلخواه دیگری برابر است.
        پس ترتیب شیر آمدن مهم نیست. 
        به محض اینکه بدانیم
        $k$
        بار شیر آمده است، 
        تمام حالات ممکن این رخداد هم‌احتمال هستند.
     </font>
</div>

<div id="sec_conditional_continuous" style="direction:rtl;line-height:300%;">

توزیع شرطی
            (حالت پیوسته)
        </font>
		<p></p>
		<hr>
 به طور مشابه با حالت گسسته، در حالت توزیع پیوسته، توزیع شرطی به شکل زیر تعریف می‌شود:
        $$f_{X|Y}(x|y) = \frac{f(x,y)}{f_Y(y)}$$
        برای شهود بهتر به توزیع نرمال دو متغیره توجه کنید. 
        می‌خواهیم ببینیم وقتی که یکی از متغیر‌ها مقدار خاصی مثلا
        $x$ را 
        اختیار می‌کند، توزیع متغیر دیگر چگونه است.
        لذا باید صفحه
        $X=x$
        را مجسم کنیم و وضعیت 
        $Y$ 
        را بر روی این صفحه نگاه کنیم. این وضعیت می‌تواند توزیع  
        $Y$
        را به ما بدهد. 
        این تقسیم کردن در فرمول توزیع شرطی را می‌توان به چشم محدود کردن فضا نگاه کرد.
        </font>
</div>

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the iris dataset
iris = load_dataset("iris")

# Filter data where Sepal.Length is between 5 and 6
dat_selected = iris[(iris["sepal_length"] > 5) & (iris["sepal_length"] < 6)]

# Plot density for Petal.Length in the filtered dataset
sns.set_theme(style="whitegrid")
plt.figure(figsize=(6, 4))

sns.kdeplot(dat_selected["petal_length"], fill=True, color="blue")

# Formatting
plt.title("Petal Length with Sepal Length Between 5 and 6")
plt.xlabel("Petal Length")
plt.ylabel("Density")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the iris dataset
iris = load_dataset("iris")

# Create the plots
sns.set_theme(style="whitegrid")

# Plot 1: Jointplot for Sepal.Length vs Petal.Length
p1 = sns.jointplot(data=iris, x="sepal_length", y="petal_length", kind="scatter", marginal_kws=dict(fill=True))

# Plot 2: Density plot for Petal Length
p2 = sns.kdeplot(iris["petal_length"], fill=True, color="red")

# Plot 3: Conditional density plot for Petal Length with Sepal Length between 5 and 6
dat_selected = iris[(iris["sepal_length"] > 5) & (iris["sepal_length"] < 6)]
p3 = sns.kdeplot(dat_selected["petal_length"], fill=True, color="blue")

# Set up the subplot grid
fig, axs = plt.subplots(1, 3, figsize=(21, 7))  # Create a grid of 1 row and 3 columns

# Plot each individual plot into the grid
p1.ax_joint.plot([], [])  # We need to remove axes for jointplot as it's a separate figure
axs[0].axis("off")
p1.figure.set_facecolor('white')
p1.figure.canvas.draw()
fig.figimage(p1.fig.canvas.renderer.buffer_rgba(), 0, 0, zorder=10)

axs[1].set_title("Total Distribution")
sns.kdeplot(iris["petal_length"], fill=True, color="red", ax=axs[1])

axs[2].set_title("Conditional Density")
sns.kdeplot(dat_selected["petal_length"], fill=True, color="blue", ax=axs[2])

# Show the plot
plt.tight_layout()
plt.show()
```

        توزیع
        $Y|X$
        در یک دیسک با توزیع یکنواخت به شکل زیر قابل محاسبه است.
        $$f(x,y) = \begin{cases}
        \frac{1}{\pi} & y^2 <= 1 - x^2 \\
        0 & o.w
        \end{cases}$$
        <br></br> 
        حل:
        ابتدا تابع چگالی مارجینالایز شده روی  
        $X$
        را پیدا می‌کنیم:
        $$f_X(x) = \int_{-\sqrt{1 - x^2}}^{\sqrt{1 - x^2}} \frac{1}{\pi} dy = \frac{2}{\pi} \sqrt{1-x^2}$$
        <br></br>
        در نتیجه:
        <br></br>
        $$f_{Y|X}(y|x) = \frac{1}{2\sqrt{1 - x^2}}$$
        </font>

        &ensp;&ensp;&ensp;&ensp;<font color="red"> مثال:</font>
        اگر تابع چگالی توام 
        $X$
        و 
        $Y$
        به صورت زیر باشد:
        $$f(x,y) = \begin{cases}
        \frac{e^{-x/y}e^{-y}}{y} & x > 0, y > 0 \\
        0 & o.w
        \end{cases}$$
        حال 
        $P(X>1|Y=y)$
        را پیدا کنید.
        به بیان دیگر اگر فرض کنیم، 
        $Y$مقداری مانند
        $y$ اختیار کرده است و خود را به آن محدود کنیم،
        توزیع 
        $X>1$
        چگونه خواهد بود؟
        <br></br>
        حل: 
        ابتدا تابع چگالی شرطی 
        $X$
        به شرط 
        $Y=y$
        را پیدا می‌کنیم:
        $$f_{X|Y}(x|y) = \frac{f(x,y)}{f_Y(y)} = \frac{e^{-x/y}e^{-y}/y}{e^{-y}\int_0^{+\infty}\frac{1}{y}e^{-x/y} dx} = \frac{1}{y} e^{-x/y}$$
        دقت کنید که این توزیع شرطی به مقدار 
        $y$
        بستگی دارد. حال می‌توان احتمال مورد نظر را حساب کرد:
        $$P(X>1|Y=y) = \int_1^{\infty} \frac{1}{y} e^{-x/y} dx = e^{-1/y}$$
        </font>
