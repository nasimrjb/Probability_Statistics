<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
آندری مارکوف که در دانشگاه پترزبورگ تحصیل می‌کرد جوان تنبل اما متفکری بود. سوالی ذهن او را مشغول کرده بود. او می‎دانست که انسان‎ها به طور میانگین 60 سال عمر می‎کنند. او می‎خواست بداند که با چه احتمالی 300 سال عمر خواهد کرد. در همین روزها بود که استادش، چبیشف، او و دوستش امینف! را به دفتر خود فراخواند تا از آن‌ها زهرچشمی بگیرد. استاد چبیشف نامساوی زیر را جلوی آن‌ها گذاشت و از آن‌ها خواست درستی یا نادرستی آن را اثبات کنند. 
<br/><b>
اگر $X$ یک متغیر تصادفی <u>نامنفی</u> و $a$ یک عدد مثبت دلخواه باشد، داریم:
<br/></b>
<br/>
    </font>
    <font size = 5>
    <center>
$$
P(X\geq a) \leq \frac{E[X]}{a} 
$$
    </center>
    <br/>
    </font>
    <font face="XB Zar" size=5>
        مارکوف به این نامساوی نگاه کرد و با خود فکر کرد:«اگر این نامساوی درست باشد، روشی برای محاسبه احتمال اینکه او  بیشتر از 300 سال عمر کند پیدا شده است.» او کاغذی پیدا کرد و روی آن نوشت:
        <br><br>
        $$
        P\{age \ge 300\} \le \frac{60}{300} = \frac{1}{5}
        $$
        <br>
        مارکوف از این نتیجه تعجب کرد اما شهودی درباره درستی این نامساوی بدست آورد. او تا به حال کسی را ندیده بود که 300 سال عمر کند اما این حقیقت درستی نامساوی را نقض نمی‎کرد. تنها نشان می‎داد کران بالایی که نامساوی ارائه می‎دهد ممکن است فاصله زیادی تا مقدار واقعی احتمال داشته باشد.
    </font>
</div>

<p></p>
<br />
<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بررسی مسئله
        </font>
		<p></p>
		<hr>
آن‌ها تصمیم گرفتند این نامساوی را روی تعدادی متغیر تصادفی دلخواه امتحان کنند تا حدس‌هایی درمورد روش اثبات یا رد آن بیابند. از آن‎جا که امینف با کتاب‌خانه‌ی seaborn آشنا بود،‌ تصمیم گرفتند به صورت گرافیکی این‌کار را انجام دهند. 
مارکوف کار با یک تاس را پیشنهاد داد. آن‌ها تصمیم گرفتند مقدار امید ریاضی در نامساوی را ثابت نگه دارند و واریانس را تغییر دهند. سه تاس شش وجهی با اعداد متفاوت را در نظر گرفتند:‌
<br/>
        <ul>
            <li>
                اعداد روی تاس اول 
                $ (1, 2, 3, 4, 5, 6) $
            </li>
            <li>
                اعداد روی تاس دوم
                $ (2.25, 2.75, 3.25, 3.75, 4.25, 4.75) $
            </li>
            <li>
                اعداد روی تاس سوم
                $ (2.875, 3.125, 3.375, 3.625, 3.875, 4.125) $
            </li>
        </ul>
نمودار احتمالِ 
        $ P(X \ge a) $
        و 
        $ \frac{mean = 3.5}{a} $
        بر حسب a رسم شده است:
</div>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dice face values
x = np.concatenate([np.arange(1, 7, 1), np.arange(2.25, 5, 0.5), np.arange(2.875, 4.25, 0.25)])

# Dice labels
facet = ["dice1"] * 6 + ["dice2"] * 6 + ["dice3"] * 6

# P(X >= a)
y = np.tile([1, 5/6, 4/6, 3/6, 2/6, 1/6], 3)

# E[X]
m = 3.5

# E[X] / a function
def f1(a):
    return m / a

# Create DataFrame
df = pd.DataFrame({"a": x, "y": y, "facet": facet})

# Plot setup
sns.set_theme(style="whitegrid")
g = sns.FacetGrid(df, col="facet", sharex=True, sharey=True, height=4)
g.map_dataframe(sns.scatterplot, x="a", y="y", hue="facet", palette="tab10", legend=False)

# Overlay E[x] / a function
a_vals = np.linspace(0.1, 6, 100)
f1_vals = f1(a_vals)

for ax in g.axes.flat:
    ax.plot(a_vals, f1_vals, color="red", linestyle="dashed", label="mean/x")
    ax.set_ylim(0, 2)
    ax.set_xlim(0, 6)
    ax.set_ylabel("P(X ≥ a)")
    ax.legend()

plt.show()
```

    </font>
</div>

<div id="sec_math" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    <font color=#FF7500 size=6>
اثبات نابرابری مارکوف
        </font>
        <hr>
         امینف که در آن مدت با کاغذ و قلم خود ورمیرفت،‌ ایده خلاقانه‌ای به ذهنش رسید! رو به مارکوف گفت:«ما می‌دانیم که اگر تابع f در تمام نقاطی که متغیر تصادفی X در آن تعریف شده کوچکتر مساوی تابع g باشد آنگاه امید ریاضی آن نیز بر روی X کمتر از g است. به عبارتی: 
<br><br>
    $$ \forall x : f(x) \leq g(x) \Rightarrow E[f(X)] \leq E[g(X)] $$
<br>
   حال تابع مشخصه‌ای را در نظر بگیر که برای مقادیر بزرگ‌تر از a‌ دلخواه، یک و مقادیر کوچک‌تر صفر باشد.    از آنجا که برای مقادیر $x$ غیر منفی این تابع کوچکتر مساوی تابع 
    $ \frac{x}{a}$
است، پس:
<br><br>
    $$ E[I_a(X)] \leq E[\frac{X}{a}] \Rightarrow P(X \geq a) \leq \frac{E[X]}{a}$$
<br>
    (چرا؟)
    مارکوف داد زد:‌«‌ آفرین!» و به سرعت روی کاغذ نوشت: 
    <br><br>
    $$
    \begin{align*}
    E[X] &= \int^{+\infty}_{0}xf(x) dx \\
    &= \int^{a}_{0}xf(x) dx + \int^{+\infty}_{a}xf(x)dx \\
    &\geq \int^{+\infty}_{a}xf(x)dx \\
    &\geq \int^{+\infty}_{a}af(x)dx \\
    &= aP(X \geq a) \\ \\
    &\Rightarrow P(X \geq a) \leq \frac{E[X]}{a}
    \end{align*}
    $$
</div>

<div id="sec_gen" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#FF7500 size=6>
تلاشی برای تعمیم
        </font>
<hr>
    مارکوف از امینف خواست تا با تست کردن چند توزیع نرمال با میانگین یکسان، ببیند شرط نامنفی بودن متغیر تصادفی X برای این نامساوی لازم است یا خیر.
امینف کد زیر را زد:
    </font>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Expected value
m = 3.5

# Function E[X] / a
def f1(a):
    return m / a

# X values
x_vals = np.linspace(0.1, 30, 300)

# Compute survival function (1 - CDF) for different standard deviations
sd_values = [15, 5, 30]
colors = ["blue", "green", "purple"]
labels = [f"sd = {sd}" for sd in sd_values]

plt.figure(figsize=(8, 5))

for sd, color, label in zip(sd_values, colors, labels):
    plt.plot(x_vals, stats.norm.sf(x_vals, loc=3.5, scale=sd), color=color, label=label)  # sf = 1 - cdf

# Plot E[X]/a function
plt.plot(x_vals, f1(x_vals), color="red", linestyle="dashed", label="mean/x")

# Formatting
plt.ylim(0, 1)
plt.xlim(0, 30)
plt.xlabel("x")
plt.ylabel("P(X ≥ x)")
plt.legend()
plt.grid(True)

plt.show()
```

    </font>

    <font face="XB Zar" size=5 color=#FF00FF>

        واریانس از چه عددی بزرگ‌تر شود، نامساوی دیگر برقرار نخواهد بود؟

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size = 6 color=#FF7500>
        نابرابری چبیشف
    </font>
    <hr/>
    <font face="XB Zar" size = 5>
    آن‌ها نتایج خود را به چبیشف نشان دادند. او که لبخندی بر لب داشت، آن‌ها را تشویق کرد و حکیمانه گفت:«نکته اصلی در نامساوی‌ای که اثباتش کردید این است که بدون دانستن pdf یک توزیع، می‌توان کرانی بر روی CDF‌ آن به دست آورد.»
    از طرفی مارکوف از نامساوی‎ای که اثبات شده بود، راضی نبود. او کرانی دقیق‎تر برای احتمال عمر 300 ساله‎اش! می‎خواست. این جا بود که از استاد چبیشف پرسید: "آیا نمی‎توان این نابرابری را دقیق‎تر کرد؟"
    <br/>
        استاد چبیشف گفت:«
        اگر به جای متغیر تصادفی 
        $X$،
        $(E[X]-X)^2$
        و به جای $a$،
        $a^2$
        را قراردهیم آنگاه به دست می‎آید: 
<br><br>
        $$
        P(|X-E[X]| \geq a) \leq \frac{Var(X)}{a^2}
        $$
<br>
این نامساوی به نام <b>چبیشف</b>، یعنی نام من مشهور است.»
<br/>
    </font>
</div>

<div style="direction:rtl; line-height:300%" align=right>
    <font face="XB Zar" size=6 color=#0000FF>
        تمرین
    </font>

    <font face="XB Zar" size=5>
        <ol>
            <li>
                بدون استفاده از نابرابری مارکوف، نابرابری چبیشف را ثابت کنید.
            </li>
            <li>
                به مارکوف کمک کنید تا با استفاده از نابرابری چبیشف، کران بالای دقیق‎تری برای عمر 300 ساله‎اش پیدا کند.
            </li>
        </ol>
    </font>

<div id="sec_other_ine" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#FF7500 size=6>
دیگر نابرابری‎ها
        </font>
    <hr>
         استاد چبیشف برخی اوقات در شرایطی قرار می‎گیرد که می‎خواهد احتمالاتی از نوعِ$P\{ X - \mu \ge a \}$
        را تخمین بزند. در این نوع از احتمالات a یک مقدار مثبت است و ما احتمالا فقط میانگین و یا واریانس X را در اختیار داریم. راه اولی که برای حل این مسئله پیشنهاد شد، استفاده از نابرابری چبیشف بود.
        $$
        \begin{align*}
        P\{ X - \mu \ge a \} \le P\{ |X - \mu| \ge a \} \le \frac{\sigma^2}{a^2} && (a > 0)
        \end{align*}
        $$
        اما استاد چبیشف ناراضی به نظر می‎رسید. بنابراین تلاش کرد تا نامساوی قوی‌تری را ارائه دهد. او در ابتدا «نابرابری یک‌طرفه‌ی چبیشف» را معرفی کرد.
        <br/>
        <b>
        اگر 
        $ X $
        یک متغیر تصادفی با میانگین صفر و واریانس متناهی باشد، در آن صورت برای هر 
        $ a \gt 0 $
        داریم:
        </b>
        $$
        P\{X \ge a\} \le \frac{\sigma^2}{\sigma^2 + a^2}
        $$
        <br/>
برای اثبات این نامساوی اگر 
        $
        b
        $
        را مثبت در نظر بگیریم، می‎توانیم بنویسیم:
        <br>
        $$
        \begin{align*}
        X \ge a && \equiv && X + b \ge a + b
        \end{align*}
        $$
        <p></p>
        در نتیجه،
        $$
        \begin{align*}
        P\{ X \ge a \} &= P\{X + b \ge a + b\} \\
        &\le P\{(X + b)^2 \ge (a+b)^2 \}
        \end{align*}
        $$
        <br>
        حال اگر نابرابری مارکوف را اعمال کنیم خواهیم داشت:
        <br><br>
        $$
        P\{X \ge a\} \le \frac{E[(X+b)^2]}{(a+b)^2} = \frac{\sigma^2 + b^2}{(a+b)^2}
        $$
        <br>
        اگر به جای
        $
        b
        $
        بگذاریم
        $
        \frac{\sigma^2}{a}
        $
        ،نابرابری یک طرفه چبیشف ثابت می‎شود. نکته قابل توجه این که 
        $
        \frac{\sigma^2}{a}
        $
        مقداری برای 
        $
        b
        $
        است که به ازای آن 
        $
        \frac{\sigma^2 + b^2}{(a+b)^2}
        $
        کمترین مقدار خود را می‎گیرد.
         به این ترتیب استاد چبیشف نابرابری یک طرفه خود را ثابت کرد.
        <br><br>
        بعد از این اثبات، استاد به این موضوع اشاره کرد که اگر میانگین
        $ X $
        صفر نباشد، در آن صورت می‎توان در نابرابری به جای 
        $ X $
        از 
        $ X - \mu $
        استفاده کرد. یعنی:
        $$
        \begin{align*}
        P\{X - \mu \ge a\} \le \frac{\sigma^2}{\sigma^2 + a^2} \\
        \Rightarrow P\{X \ge a + \mu\} \le \frac{\sigma^2}{\sigma^2 + a^2}
        \end{align*}
        $$
</font>
</div>
