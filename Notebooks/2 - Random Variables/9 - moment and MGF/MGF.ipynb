<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مقدمه
        </font>
		<p></p>
		<hr>
        در مطالعه‌ی متغیرهای تصادفی، علاوه بر معیارهای پایه‌ای مانند امید ریاضی و واریانس، ابزارهای دیگری برای توصیف توزیع یک متغیر تصادفی مورد استفاده قرار می‌گیرند. یکی از این ابزارها گشتاورها (Moments) هستند که اطلاعاتی درباره‌ی شکل، پراکندگی و رفتار توزیع تصادفی ارائه می‌دهند.
        <br>
        به طور کلی، گشتاورهای یک متغیر تصادفی، میانگین توانی از متغیر تصادفی هستند و می‌توانند به ما کمک کنند تا خصوصیات مهم توزیع را بهتر تحلیل کنیم. برای مثال، میانگین و واریانس که از پرکاربردترین شاخص‌های آماری هستند، در واقع گشتاورهای مرتبه اول و دوم محسوب می‌شوند. اما بررسی گشتاورهای مرتبه بالاتر، اطلاعات بیشتری در مورد توزیع به ما می‌دهد، مانند چولگی (Skewness) و کشیدگی (Kurtosis) که به ترتیب عدم تقارن و پهنای توزیع را نشان می‌دهند.
        <br>
        برای محاسبه و تحلیل گشتاورها، یک ابزار مفید به نام تابع مولد گشتاور (Moment Generating Function - MGF) معرفی می‌شود. MGF نه‌تنها گشتاورهای توزیع را محاسبه می‌کند، بلکه در حل بسیاری از مسائل آماری و نظریه‌ی احتمالات، از جمله تعیین توزیع‌های حدی و قضایای همگرایی، کاربرد دارد.

<div id="sec_nth_moment" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            گشتاور های مرتبه مختلف
        </font>
		<p></p>
		<hr>
        گشتاورها توابعی از متغیر تصادفی هستند که اطلاعاتی درباره‌ی شکل و ویژگی‌های توزیع ارائه می‌دهند. به‌طور کلی، گشتاور مرتبه‌ی $k$ یک متغیر تصادفی $X$ را به صورت زیر تعریف میکنیم
        $$E[X^k]$$
        که $k$ ذر آن یک عدد صحیح مثبت است.
        <br>
        گشتاور مرکزی مرتبه $k$ متغیر $X$ را نیز به این صورت تعربف میکنیم.
        $$ E[(X - E[X])^k]$$
        با کمی دقت متوجه میشویم که برای مثال گشتاور مرتبه یک $$E[X]$$ همان امید ریاضی و گشتاور مرکزی مرتبه دو $ E[(X - E[X])^2]$ همان واریانس هست.
        <br>
        به همین ترتیب دیگر گشتاور ها نیز اطلاعات زیادی راجع به متغیر تصادفیمان به ما میدهند
        <br>
        

</font>
</div>

<div id="sec_relation" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            ارتباط بین گشتاورها و ویژگی‌های توزیع
        </font>
		<hr>
        میانگین: گشتاور مرتبه اول
		<br>
		$$ \mu = E[X] = \sum_{x} x P(X = x) $$
		<br>
		واریانس: گشتاور مرکزی مرتبه دوم
		<br>
		$$ \sigma^2 = E[(X - \mu)^2] = E[X^2] - (E[X])^2 $$
		<br>
		چولگی: گشتاور مرکزی مرتبه سوم
		<br>
		$$ \text{Skewness} = \frac{E[(X - \mu)^3]}{\sigma^3} $$
		<br>
		کشیدگی: گشتاور مرتبه چهارم
		<br>
		$$ \text{Kurtosis} = \frac{E[(X - \mu)^4]}{\sigma^4} $$
		<br>
		


</font>
</div>

<div id="sec_MGF_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            تعریف تابع مولد گشتاور(MGF)
        </font>
		<p></p>
		<hr>
		تابع مولد گشتاور (Moment Generating Function - MGF) یکی از ابزارهای مهم در آمار و احتمال است که اطلاعات مربوط به توزیع یک متغیر تصادفی را خلاصه می‌کند. این تابع به‌صورت زیر تعریف می‌شود:
		$$ M_X(t) = E[e^{tX}] $$
		که در این صورت میگوییم تابع مولد گشتاور متغیر $x$ وجود دارد اگر که مقدار ثابت a وجود داشته باشد به طوری که $M_X(t)$ برای تمام مقادیر
		$ t \in [-a, a]$
		وجود داشته باشد.
		<br>
		<br>
		با توجه به تعریف $M_X(t)$ داریم برای متغیر گسسته:
		<br>
		$$ M_X(t) = E[e^{tX}] = \sum_{x} e^{tX} P(X = x) $$
		<br>
		و برای متغیر تصادفی پیوسته:
		<br>
		$$ M_X(t) = E[e^{tX}] = \int_{-\infty}^{\infty} e^{tX} f(x) \,dx $$
		<br>

</font>
</div>

<p></p>
<br />
<div id="sec_MGF_moment" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            رابطه بین MGF و گشتاورها
        </font>
		<hr>
        از ویژگی های مهم تابع مولد گشتاور این است که میتوان تمام گشتاورهای متغیر تصادفی را از روی آن استخراج کرد. این کار از طریق مشتق‌گیری از $M_X(t)$ بدست آورد.
        <br>
        می دانیم که بسط تیلور برای $e^x$ به این صورت است
        <br>
        $$ e^x = \sum_{k=0}^{\infty} \frac{x^k}{k!} = 1 + \frac{x}{1!} + \frac{x^2}{2!} + \frac{x^3}{3!} + \dots $$
        <br>
        که از این میتوان رسید به:
        <br>
        $$ e^tX = \sum_{k=0}^{\infty} \frac{(tX)^k}{k!} = \sum_{k=0}^{\infty} \frac{X^k t^k}{k!}$$
        <br>
        و درنهایت میتوان به این فرمول برای محاسبه گشتاور ها از MGF رسید:
        <br>
        $$ M_X(t) = E[e^{tX}] = \sum_{k=0}^{\infty} E[X^k] \frac{t^k}{k!} $$
        <br>
        پس همانطور که دیدید میتوان با داشتن تابع مولد گشتاور متغیر تصادفی $X$ تمام گشتاور های آن را بدست آورد.
        <br>
        که برای اینکار کافیست از $M_x(t)$ نسبت به $t$، $k$ بار مشتق گرفت تا گشتاور مرتبه $k$ آن را بدست آورد.
        <br>
        $$ E[X^k] = M_X^{(k)}(0) = \left. \frac{d^k}{dt^k} M_X(t) \right|_{t=0} $$

</font>
</div>

<div id="sec_MGFs" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            تابع مولد گشتاور برای توزیع های مختلف
        </font>
		<hr>
		با استفاده از فرمول گفته شده میتوان MGF  را برای توزیع های پرکاربرد محاسبه کرد.
		<br>
		<br>
		توزیع برنولی (Bernoulli):
		$$ M_X(t) = (1 - p) + p e^t $$
		<br>
		توزیع دو جمله‌ای (Binomial):
		$$ M_X(t) = (1 - p + p e^t)^n $$
		<br>
		توزیع پواسون (Poisson):
		$$ M_X(t) = e^{\lambda(e^t - 1)} $$
		<br>
		توزیع نرمال (Normal):
		 $$ M_X(t) = e^{\mu t + \frac{1}{2} \sigma^2 t^2} $$
		<br>
		توزیع نمایی (Exponential):
		$$ M_X(t) = \frac{\lambda}{\lambda - t}, \quad t < \lambda $$



        

</font>
