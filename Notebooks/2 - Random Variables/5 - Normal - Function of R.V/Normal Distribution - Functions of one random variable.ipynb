            توزیع نرمال
        </font>
		<p></p>
        <hr>
توزیع نرمال یا گوسی مهم ترین توزیع احتمالاتی در آمار میباشد زیرا این توزیع با بسیاری از پدیده های طبیعی تطابق پیدا میکند. به عنوان مثال، قد افراد، فشار خون، خطا های اندازه گیری، و میزان IQ افراد از مواردی هستند که از این توزیع پیروی میکنند.
<br>
	</font>
</div>

<p></p>
<br />
<div id="normal_attributes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            ویژگی ها
        </font>
		<p></p>
        <hr>
تابع چگالی احتمال توزیع نرمال با پارامترهای $\mu$ و $\sigma ^{2}$ به صورت زیر است :
        <br>
        $$ f(x; \mu,\sigma ^{2}) = \frac{1}{\sqrt{2\pi\sigma^{2}}} e ^{\frac{-(x-\mu) ^ {2}}{2\sigma ^{2}}}$$
        <img src="figs/Normal_Distribution_PDF.png" />
        <br>
        امید ریاضی و واریانس آن هم همان پارامتر هایش می‌باشد:
        $$ E[X] = \mu $$
        $$ Var(X) = \sigma ^{2} $$
        <font size=4>
        (اثبات آن را در کتاب راس صفحه 200 میتوانید مشاهده کنید.)
        </font>
        <br>
        <br>
        اگر $X \sim N(\mu,\sigma ^{2})$ و a و b دو عدد حقیقی باشند؛ آنگاه:
        $$ aX + b \sim N(a\mu + b, (a\sigma)^{2}) $$
        <br>
        توزیع نرمال با امید ریاضی 0 و واریانس 1 را توزیع نرمال استاندارد میگویند. آن را با Z نشان میدهیم.
        تابع توزیع تجمعی(CDF) در توزیع نرمال استاندارد با $\phi$ نشان داده میشود.
        $$ F(x) = \phi(x) $$
        تابع توزیع تجمعی فرمول تمیزی ندارد و از جدول $\phi$ برای آن استفاده میکنیم.
        <br>
        <br>
        در حالت کلی برای متغیر تصادفی X با توزیع نرمال $N(\mu,\sigma^{2})$ طبق فرمول ترکیب خطی داریم:
        $$ Z = \frac{X - \mu}{\sigma} $$
        $$ F(x) = P(X \le x) = P(\frac{X - \mu}{\sigma} \le \frac{x - \mu}{\sigma}) = \phi(\frac{x - \mu}{\sigma}) $$
        <br>
        <img src="figs/phi_table.png" />
        برای اعداد منفی هم با توجه به تقارن در توزیع نرمال استاندارد داریم:
        $$ \phi(-x) = 1 - \phi(x) $$
<br>
	</font>
</div>

<br />
<div id="normal_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مثالی از توزیع نرمال
        </font>
		<p></p>
        <hr>
        اگر X یک متغیر تصادفی نرمال با پارامتر های $\mu = 3$ و $\sigma^{2} = 9$ باشد احتمال $P(2 < X < 5)$ چقدر است؟
        <br>
        <br>
        $$ P(2 < X < 5) = P(\frac{2 - 3}{3} < \frac{X - 3}{3} < \frac{5 - 3}{3}) $$
        $$ = P(-\frac{1}{3} < Z < \frac{2}{3}) $$
        $$ = \phi(\frac{2}{3}) - \phi(-\frac{1}{3}) $$
        $$ = \phi(\frac{2}{3}) - [1 - \phi(\frac{1}{3})] $$
        $$ \approx 0.7486 - [1 - 0.6293] = 0.3779 $$
<br>
	</font>
</div>

        </font>

        اردشیر به تازگی منشی یک شرکت  شده است. به او گفته شده است که برای حفظ پرستیژ شرکت تازه تاسیس، تعداد افرادی که در هر ساعت به کار آن‌ها رسیدگی می‌شود باید از توزیع یکنواخت بین m تا n نفر

<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مقدمه
        </font>
		<p></p>
        <hr>
        در قسمت‌های قبلی با متغیر تصادفی آشنا شدیم. می‌دانیم متغیر تصادفی یک نگاشت از فضای نمونه به اعداد حقیقی است. 
        $$ X: \Omega \rightarrow \mathbb{R} $$
        <img src="figs/RV.png" style="float:center; width:40%"/>
        <br>
        حال می‌خواهیم با تغییر (transformation)  های متغیرهای تصادفی آشنا شویم. که ساده‌ترین آن‌ها، تابعی از یک متغیر تصادفی است. 
        <br>
        Y را به‌صورت تابعی از متغیر تصادفی پیوسته‌ی X به‌شکل زیر تعریف می‌کنیم.
        <br>
        $$ \textbf{Y} = g(\textbf{X}) $$
        $$ g: \mathbb{R} \rightarrow \mathbb{R} $$
        <br>
        که در واقع خلاصه‌شده‌ی تعریف زیر است:
        $$  \textbf{Y} (\omega) = g(\textbf{X}(\omega)) , \ \omega \in \Omega $$
        <br>
        در این‌جا، Y نیز یک متغیر تصادفی پیوسته خواهد بود و توزیع خاص خود را خواهد داشت. 
        <br>
        <img src="figs/y_omega.png" style="float:center; width:50%"/>
        <br> <br>
        در این نوت‌بوک دو روش برای به‌دست آوردن تابع چگالی متغیر تصادفی Y معرفی می‌کنیم. اول، روی g های یک به یک متمرکز شده و پس از آن بقیه توابع را بررسی خواهیم کرد. در آخر نیز یاد می‌گیریم چطور با استفاده از معکوس توابع توزیع تجمعی، می‌توان از هر توزیعی که در دسترس داریم، توزیع دیگری را بسازیم.
	</font>
</div>

<p></p>
<br/>
<div id="sec_CDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
        به‌دست آوردن تابع توزیع تجمعی
        </font>
        <hr>
        برای به دست آوردن توزیع $ Y = g(X) $ می‌توان به‌صورت مستقیم و با استفاده از تعریف تابع توزیع تجمعی، تابع توزیع تجمعی این متغیر تصادفی را به‌دست آورد. یعنی به طور کلی به‌صورت زیر عمل کرد:
        <br> 
$$ F_Y(y) = P(Y \le y) = P(g(X) \le y) = \sum P(x_i \le X \le x_j) = \sum \pm \ F_X(x_i) $$
        <br>
در حالت خاص اگر تابع g یک به یک و صعودی باشد خواهیم داشت:
        <br>
$$ F_Y(y) = P(Y \le y) = P(g(X) \le y) = P(X \le g ^{-1} (y)) = F_X(g ^{-1} (y)) $$
        <br>
        پس برای تابع توزیع چگالی Y در این حالت خواهیم داشت:
        <br>
        $$ f_Y(y) = \frac{d F_Y(y)}{dy} =\frac {d F_X(g ^{-1} (y))}{dy} = \frac{f_x(g ^{-1} (y))}{g'(g ^{-1} (y))} $$
        <br>
        محاسبات مربوط به حالت کلی در بخش بعد آمده است. 
        <br>
    <hr>
    <font color=blue>
            مثال ۱ (تابع خطی): 
    </font>
    فرض کنید X یک متغیر تصادفی پیوسته با توزیع معلوم است. توزیع Y را به‌دست آورید. 
    $$ Y = aX + b $$
    <br> 
    با استفاده از روش معرفی شده در بالا به‌صورت زیر عمل می‌کنیم.
    <br>
    $$ \begin{align*}
    F_Y(y) &= P(Y \le y) = P(aX+b \le y)  \\
    &\overset{a > 0}{=} P(X \le \frac{y - b}{a}) = F_X(\frac{y - b}{a})  \\
    &\overset{a < 0}{=} P(X \ge \frac{y - b}{a}) = 1 - F_X(\frac{y - b}{a})
    \end{align*} $$
    <br>
    <hr>
    <font color=blue>
            مثال ۲:
    </font>
    فرض کنید X یک متغیر تصادفی پیوسته با توزیع معلوم است. توزیع Y را به‌دست آورید. 
    $$ Y = X^2  $$
    <br> 
    همانند مثال قبل داریم:
    <br>
    $$ F_Y(y) = P(Y \le y) \overset{y > 0}{=} 
P(-\sqrt{y} \le X \le \sqrt{y}) = F_X(\sqrt{y}) - F_X(- \sqrt{y}) $$
    <br>
    همچنین می‌توان تابع توزیع چگالی را به‌روش زیر به‌دست آورد:
    <br>
    $$ \frac{d F_X(\sqrt{y})}{dy} = \frac{1}{2\sqrt{y}} f_X(\sqrt{y}) \Rightarrow
     f_Y(y) = \frac{1}{2\sqrt{y}} f_X(\sqrt{y}) + \frac{1}{2\sqrt{y}} f_X(-\sqrt{y}) , \ (y>0) $$
    <br> 
    <img src="figs/X2.png" style="float:center; width:80%"/>
	</font>
</div>

<div id="sec_PDF" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                به‌دست آوردن تابع توزیع چگالی به‌صورت مستقیم
        </font>
        <hr>
        <font color=#FF7500>
                قضیه:
        </font>
        اگر X یک متغیر تصادفی باشد و متغیر تصادفی Y از رابطه‌ی $Y=g(X)$ به‌دست آمده باشد
و
$x_1, x_2, … $ 
ریشه‌های معادله‌ی $y=g(x)$ باشند. یعنی:
$ y = g(x_1) = … = g(x_i) = … $
        <br>
        تابع توزیع چگالی Y در نقطه‌ی y از رابطه‌ی زیر به‌دست می‌آید.
        <br>
        <br>
        $$ f_Y(y) = \frac{f_X(x_1)}{|g'(x_1)|} + ... + \frac{f_X(x_i)}{|g'(x_i)|} + ... = \sum_i \frac{f_X(x_i)}{|g’(x_i)|} $$
        <br>
        (اثبات قضیه را می‌توانید در صفحه‌ی ۱۱۸، بخش ۴-۴ منبع اول پیدا کنید.)
    <br>
    <img src="figs/pdf_theorem.png" style="float:center; width:50%"/>
    <br>
    <br>
    <font color="blue">
            مثال:
    </font>
فرض کنید یک لیزر را که در فاصله‌ی واحد از مبدا قرار دارد، (مطابق شکل) به زمین می‌تابانیم. متغیر تصادفی $X$ بیانگر فاصله افقی نقطه‌ی برخورد این پرتو نسبت به مبدا در آزمایش‌های متوالی است. اگر بدانیم 
$\theta$
که بیانگر زاویه قرار گیری لیزر است از توزیع یکنواخت 
$(- \pi / 2, \pi / 2)$
پیروی می‌کند، بررسی کنید $X$ از چه توزیعی است. 
    <img src="figs/cauchy.png" style="float:center; width:50%"/>
     حل: ابتدا متغیر تصادفی $X$ را به‌صورت تابعی از متغیر تصادفی
     $\theta$
     که توزیع آن را داریم می‌نویسیم:
     $$ X = tan(\theta)$$
   حال با استفاده از روش بالا خواهیم داشت:
   $$ f_X(x) = \frac{f_{\theta} (\theta _0)}{1 + tan^2 \theta_0} , \;\; \theta_0 = tan^{-1} x 
   \Rightarrow f_X(x) = \frac{\frac{1}{\pi}}{1 + x^2} =  f_X(x) = \frac{1}{\pi(1 + x^2)} $$
	</font>
</div>

<div id="sec_exp" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امید ریاضی تابعی از متغیر تصادفی
        </font>
		<p></p>
		<hr>
 می‌دانیم در برخی از کاربردها تنها نیاز داریم امید ریاضی تابعی از یک متغیر تصادفی را محاسبه کنیم و نمی‌خواهیم توزیع آن متغیر تصادفی را به‌دست آوریم. به طور مثال محاسبه‌ی امید ریاضی مربع یک متغیر تصادفی
از مواردی است که در درس به زودی با آن مواجه خواهید شد.
<br/>
نکته‌ای که باید به آن توجه کرد این است که امید ریاضی تابعی از یک متغیر لزوماً برابر با اعمال آن تابع بر روی امید ریاضی نیست.
<br/>
به طور مثال فرض کنید متغیر تصادفی
$X$
دو مقدار
۰
و
۲
را با احتمال برابر بتواند اختیار کند. داریم:
$$
\mathop{\mathbb{E}}[X] = 0 \times \frac{1}{2} + 2 \times \frac{1}{2} = 1
$$
این در حالی است که امید ریاضی متغیر تصادفی
$Y = g(X) = X^2$
برابر است با:
$$
\mathop{\mathbb{E}}[Y] = 0^2 \times \frac{1}{2} + 2^2 \times \frac{1}{2} = 2
$$
در نتیجه
$\mathop{\mathbb{E}}[X^2] = 1^2 \neq (\mathop{\mathbb{E}}[X])^2 = 2$
و در حالت کلی نتیجه می‌شود که:
$$
g(\mathop{\mathbb{E}}[X]) \neq \mathop{\mathbb{E}}[g(X)]
$$
<br/>
در حالت کلی امید ریاضی تابع
$g$
از متغیر تصادفی
X
برابر است با
$$
\mathop{\mathbb{E}}[g(X)] = \sum_{i}{g(x_i)p(x_i)}
$$
که در آن
به ازای
$i \geq 1$،
$x_i$
ها
همه مقادیری هستند که
$X$
می‌تواند اختیار کند.
<br/>
برای اثبات رابطه بالا فرض می‌کنیم که به ازای
$i \geq 1$،
های مختلف،
مقادیر
$y_j$
ها که
$j \geq 1$
برابر با مقادیر متفاوت
$g(x_i)$
ها باشند. سپس از طریق دسته‌بندی همه
$i$
هایی که
$g(x_i)$
یکسان است،
رابطه ثابت می‌گردد.
        $$
        \begin{aligned}
        \sum_{i}{g(x_i)p(x_i)} &= \sum_{j}\sum_{i:g(x_i)=y_j}{g(x_i)p(x_i)}\\
        &= \sum_{j}\sum_{i:g(x_i)=y_j}{y_jp(x_i)}\\
        &= \sum_{j}y_j\sum_{i:g(x_i)=y_j}{p(x_i)}\\
        &= \sum_{j}y_jp\{g(X)=y_j\}\\
        &= \mathop{\mathbb{E}}[g(X)]
        \end{aligned}
        $$
<br/>
در حالت پیوسته هم رابطه زیر نتیجه می‌گردد.
$$
\mathop{\mathbb{E}}[g(X)] = \int_{-\infty}^{\infty}{g(x)f_X(x)dx}
$$
	</font>
</div>

<br />
<div id="sec_definition_example_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color="blue">
            مثال:
    </font>
تکه چوبی به طول ۱ واحد داریم. این تکه چوب را از نقطه
U
می‌شکنیم. تابع توزیع احتمال
U
به صورت یکنواخت
است یعنی:
$$
f(u) = \begin{cases}
       1 &\quad u\in [0,1]\\
       0 &\quad\text{otherwise} \\ 
     \end{cases}
$$
نقطه
$0 \leq p \leq 1, p$
به ما داده شده است.
امید ریاضی طول قطعه چوبی را که شامل نقطه
$p$
است به دست آورید.
<br/>
<img src="figs/Stick_Function_Example.png" style="float:center; width:35%"/>
<br />
تابع
$L_p(U)$
از متغیر تصادفی
U
را تعریف می‌کنیم:
$$
L_p(U) = \begin{cases}
       1 - U &\quad U < p\\
       U &\quad U > p \\ 
     \end{cases}
$$
حال برای به دست آوردن امید ریاضی این تابع داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[L_p(U)] &= \int_{-\infty}^{\infty}{L_p(u)f(u)du} \\
&= \int_{0}^{1}{L_p(u).1.du} \\
&= \int_{0}^{p}{(1-u)du} + \int_{p}^{1}{udu}\\
&= p(1-p) + \frac{1}{2}
\end{aligned}
$$
	</font>
</div>

<div id="sec_ITM" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                روش تبدیل معکوس (یا Inverse Transform Sampling)
        </font>
        <hr>
        فرض کنیم منبعی وجود داشته باشد که بتوانیم با آن نمونه‌هایی از توزیع پیوسته‌ی X با تابع توزیع تجمعی $F_X$  به‌دست آوریم. و می‌خواهیم نمونه‌هایی از توزیع مشخص Y داشته باشیم. حال می‌خواهیم تابع g رابه‌گونه‌ای به‌دست بیاوریم که بتوانیم با نمونه‌هایی از توزیع X، نمونه‌های توزیع Y را بسازیم. یعنی:
<br>
        $$ Y = g(X) $$
<br><br>
۱. فرض کنید $F$ تابع توزیع تجمعی متغیر تصادفی X باشد. می‌دانیم 
$ F: \mathbb{R} \longrightarrow [0, 1] $
بنابراین، 
$ 0 \le Y \le 1 $
حال می‌خواهیم برای Y، تابع توزیع تجمعی $F$ را به‌دست آوریم.
<br><br>
$$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (F(X) \le y) =  \mathbb{P} (X \le F^{-1}(y)) =
F_X(F^{-1}(y)) = y $$
<br>
بنابراین Y توزیع یکنواخت بین $[0, 1]$ است. 
<br><br>
۲. اگر فرض کنیم U تابع توزیع یکنواخت بین $[0, 1]$ و h یک تابع اکیداً صعودی با برد $[0, 1]R =  $  باشد. می‌خواهیم ثابت کنیم که اگر داشته باشیم:
$$ Y = h^{-1}(U) $$ 
آن‌گاه h تابع توزیع تجمعی Y است. 
<br><br>
می‌دانیم h صعودی است. بنابراین $h^{-1}$ نیز صعودی است. پس داریم:
<br>
$$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (h^{-1}(U) \le y) \overset{h^{-1}\text{ is increasing}}{=} \mathbb{P} (U \le h(y)) =  F_U(h(y)) \overset{\text{U is uniform}}{=} h(y) $$
<br><br>
۳. در نهایت فرض می‌کنیم توزیع X را داریم. برای تابع دلخواه صعودی $F_Y$ با شرط‌های  
$ lim_{x \rightarrow\infty} F_Y(x)dx = 1$ و $ lim_{x \rightarrow- \infty} F_Y(x)dx = 0 $
، می‌خواهیم تابع g در $Y = g(X)$ را به‌گونه‌ای به‌دست آوریم که این تابع، CDF متغیر تصادفی Y باشد. 
<br><br>
با استفاده از قسمت ۱ می‌دانیم توزیع U به‌صورت زیر به‌دست می‌آید:
$$ U = F_X(X) $$
<br>
همچنین از قسمت ۲ می‌دانیم:
$$ Y = F_Y^{-1} (U) $$
<br>
بنابراین خواهیم داشت:
$$ Y =  F_Y^{-1} (F_X(X)) $$
<br> <br>
می‌توان به‌روش زیر نیز تابع g را به‌دست آورد:
$$ F_Y(y) = \mathbb{P}(Y \le y) = \mathbb{P} (g(X) \le y)
= \mathbb{P} (X \le g^{-1} (y)) = F_X(g^{-1} (y))
 \Rightarrow y = F_Y^{-1}(F_X(g^{-1} (y))) 
 \Rightarrow g(y) = F_Y^{-1}(F_X(y)) $$
<br><br>
	</font>
</div>

```python
library(ggplot2)
m = 5
n = 10
lambda = 1
N = 10000
X = rexp(N, lambda)
Y = lambda * exp(-X)
Y = (n - m)*Y + m
df = data.frame(Y = Y, exp = runif(N, min=m, max=n)) # compare with uniform distr
ggplot(df) + geom_histogram(aes(exp), alpha = 0.4, fill = "dark blue", binwidth = 0.05) +
  geom_histogram(aes(Y), alpha = 0.3, fill = "red", binwidth = 0.05)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

m = 5
n = 10
lambda_ = 1
N = 10000

# Generate random samples
X = np.random.exponential(scale=1/lambda_, size=N)
Y = lambda_ * np.exp(-X)
Y = (n - m) * Y + m

# Create a DataFrame
df = pd.DataFrame({'Y': Y, 'exp': np.random.uniform(low=m, high=n, size=N)})

plt.figure(figsize=(10, 6))
sns.histplot(df['exp'], color='darkblue', alpha=0.4, binwidth=0.05, label='Uniform')
sns.histplot(df['Y'], color='red', alpha=0.3, binwidth=0.05, label='Transformed Exponential')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Distributions')
plt.show()

```
